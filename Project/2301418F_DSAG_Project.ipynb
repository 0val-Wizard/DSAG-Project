```python
# this function checks for if the input is a (palindrome)/(palindrome by word-unit)
def check_palindrome(feed, option):
    
    if option == "1":
        # check for palindrome word-unit
        cleaned_input = []

        # Remove space, punctuations and make capatilzation small
        small_character = feed.lower()
        small_space_removed = small_character.strip()
        filtered_input = ''.join(char for char in small_space_removed if char.isalnum())
    
        # Appending user_input into cleaned_input
        for ele in filtered_input:
            cleaned_input.append(ele)  
            
        # checks if input is a palindrome by slicing --> [start:stop:step]
        return cleaned_input == cleaned_input[::-1]
   

    elif option == "2":
        # check for palindrome by word-unit 
        cleaned_input=[]

        # Remove punctuations and make capatilzation small
        small_character = feed.lower()
        filtered_input = ''.join(char for char in small_character if char.isalnum()) # isalnum() checks whether the char (character) picked is alphanumeric

        # Organising user_input into words and appending the words into cleaned_input
        for ele in filtered_input.split(): # consecutive spaces considered as one seperator
            cleaned_input.append(ele)
            
        # checks if input is a palindrome by word-unit by slicing --> [start:stop:step]
        return cleaned_input == cleaned_input[::-1]
        
        
# this function asks the user if he wants to go again
def ask():
    ask_again = True
    
    while ask_again == True:
        ask = input("Try again?[y/n]:").lower().strip()
        
        if ask == "y":
            # user wants to play again
            print("PLAYING AGAIN ✅\n")
            ask_again = False
            return True
        
        elif ask == "n":
            # user does not want to play again
            print("GAME OVER\n")
            ask_again = False
            return False
        
        else:
            # input is invalid
            print("You input is INCORRECT. (❌) Type your input again")
            ask_again = True
        
        
# while loop to check for Palindrome --> Code will stop with the input received from the ask() function    
go_again=True
while go_again:
    
    # print menu/options
    print("\n\t\tFun with Palindromes")
    print("Menu\n\n")
    print("Option 1: Palinrdrome")
    print("Option 2: Palindrome by Word-unit\n")

    # type of palindrome picked by the user
    option_choosen = str(input("Option[pick a number]: ")).strip() # Remove any spaces in given input --> error-free
    
    
    if option_choosen == "1":
        # Option1: Check if user_input is a palindrome
        # takes user's input
        user_input = input("Am I a Palindrome?: ")
        result = check_palindrome(user_input, "1")
        
        if result == True:
            # user_input is a palindrome
            print(f"\nYes! '{user_input}' is a Palindrome✅\n")
            go_again = ask()
        else:
            # user_input is not a palindrome
            print(f"\nNo. '{user_input}' is not a Palindrome(❌)\n")
            go_again = ask()
            
            
    elif option_choosen == "2":
        # Option2: Check if user_input is a palindrom by word-unit
        # takes user's input
        user_input = input("Am I a Palindrome?: ")
        result = check_palindrome(user_input, "2")
        
        if result == True:
            # user_input is a palindrome by word-unit
            print(f"\nYes! '{user_input}' is a Palindrome by Word-unit✅\n")
            go_again = ask()
        else:
            # user_input is not a palindrome by wor-unit
            print(f"\nNo. '{user_input}' is not a Palindrome by Word-unit(❌)\n")
            go_again = ask()
            
    
    else:
        # Asks for option_chossen input again if input is invalid
        print("Your option input is invalid\n")
        go_again = ask()
```
